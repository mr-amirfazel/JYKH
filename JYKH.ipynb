```python
def json_reader(path):
  f = open(path)
  data = json.load(f)
  return data

def save_to_json(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

def load_from_json(file_path):
    with open(file_path, 'r', encoding='utf-8') as json_file:
        data = json.load(json_file)
    return data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import time


def process_query(query, d=5):
  start_time = time.time()
  query_embedding = model.encode(query, convert_to_tensor=False)
  similarities = []

  for i in range(len(contents)):
    file_path = f'/content/drive/MyDrive/IR_Bonus_Project/phase_1_results/{i}.json'
    loaded_data = load_from_json(file_path)
    doc_embedding = loaded_data["embedding"]
    similarity = cosine_similarity([query_embedding], [doc_embedding])[0][0]
    similarities.append({"doc_id": loaded_data["doc_id"], "content": loaded_data["content"], "similarity": similarity})

  similarities = sorted(similarities, key=lambda x: x["similarity"], reverse=True)



  print(f"Top {d} Documents for query -- \"{query}\" -- : ")
  for rank, result in enumerate(similarities[:d], start=1):
      print(f"Rank {rank}:")
      print(f"Document ID: {result['doc_id']}")
      print(f"Content: {result['content']}")
      print(f"Similarity: {result['similarity']:.4f}")
      print()

  end_time = time.time()
  print(f"Retrieval Time: {end_time - start_time:.2f} seconds\n--------------------")
```

```python
for query in queries:
  process_query(query)
```

```python
import numpy as np

def prepare_all_embeds():
  document_ids = []
  embeddings = []


  for i in range(len(contents)):
    embeddings_path = f'/content/drive/MyDrive/IR_Bonus_Project/phase_1_results/{i}.json'
    loaded_data = load_from_json(embeddings_path)
    document_ids.append(loaded_data["doc_id"])
    embeddings.append(loaded_data["embedding"])

  embeddings_matrix = np.array(embeddings)

  return embeddings_matrix, document_ids
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import time

start_time = time.time()

all_embeds, document_ids = prepare_all_embeds()

from sklearn.metrics import silhouette_score

# silhouette_scores = []
# for k in range(2, 15):  # Minimum k = 2 for silhouette score
#     kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
#     kmeans.fit(all_embeds)
#     score = silhouette_score(all_embeds, kmeans.labels_)
#     silhouette_scores.append(score)

# # Plot silhouette scores
# plt.plot(range(2, 15), silhouette_scores, marker='o')
# plt.xlabel('Number of clusters (k)')
# plt.ylabel('Silhouette Score')
# plt.title('Silhouette Method')
# plt.show()

k = 93

kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(all_embeds)

cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

clusters = {i: [] for i in range(k)}
for doc_id, label in zip(document_ids, cluster_labels):
    clusters[label].append(doc_id)

for cluster_id, doc_ids in clusters.items():

    output_path = f'/content/drive/MyDrive/IR_Bonus_Project/phase_3_results/cluster_{cluster_id}.json'
    data_to_save = {
        "documents": doc_ids,
        "centroid": centroids[cluster_id].tolist(),
        "cluster_id": cluster_id
    }
    save_to_json(data_to_save, output_path)
    print(f"Cluster {cluster_id} contains docs: {doc_ids}")

end_time = time.time()
print(f'clustering duration: {end_time - start_time}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import time

def process_query_clusters(query, clusters, all_embeddings_path):
    start_time = time.time()
    query_embedding = model.encode(query, convert_to_tensor=False)

    cluster_similarities = []
    for cluster in clusters:
        centroid = np.array(cluster["centroid"])
        similarity = cosine_similarity([query_embedding], [centroid])[0][0]
        cluster_similarities.append((cluster["cluster_id"], similarity))

    cluster_similarities = sorted(cluster_similarities, key=lambda x: x[1], reverse=True)

    results = []
    for cluster_id, cluster_similarity in cluster_similarities[:3]:
        cluster = next(c for c in clusters if c["cluster_id"] == cluster_id)
        for doc_id in cluster["documents"]:
            doc_file_path = f"{all_embeddings_path}/{doc_id}.json"
            doc_data = load_from_json(doc_file_path)
            doc_embedding = np.array(doc_data["embedding"])
            doc_content = doc_data["content"]

            doc_similarity = cosine_similarity([query_embedding], [doc_embedding])[0][0]
            results.append({
                "doc_id": doc_id,
                "content": doc_content,
                "similarity": doc_similarity,
                "cluster_id": cluster_id
            })

    results = sorted(results, key=lambda x: x["similarity"], reverse=True)
    end_time = time.time()
    duration = end_time - start_time
    return results, duration
```

```python
queries = ["ویروس کرونا",
           "خزندگان سمی استان بوشهر",
           "رشته کوه های هیمالیا",
           "لیگ قهرمانان اروپا",
           "تلسکوپ و فضا پیما ها در مریخ و زحل"
          ]
```

```python
import time


k = 93
clusters = []
for cluster_id in range(k):
    file_path = f"/content/drive/MyDrive/IR_Bonus_Project/phase_3_results/cluster_{cluster_id}.json"
    cluster_data = load_from_json(file_path)
    clusters.append(cluster_data)

all_embeddings_path = "/content/drive/MyDrive/IR_Bonus_Project/phase_1_results/"


for query in queries:
  ranked_results, duration = process_query_clusters(query, clusters, all_embeddings_path)
  print("<=============================================>")
  print(f"Top results for the query \"{query}\" retrieved in {duration} seconds:")
  for rank, result in enumerate(ranked_results[:5], start=1):
      print(f"\n\tRank {rank}:")
      print(f"\tDocument ID: {result['doc_id']}")
      print(f"\tDocument Content: {result['content']}")
      print(f"\tFound in Cluster id: {result['cluster_id']}")
      print(f"\tSimilarity: {result['similarity']:.4f}\n---------------------------------")



```

```python
import faiss
import numpy as np

embeddings_path = "/content/drive/MyDrive/IR_Bonus_Project/phase_1_results/"
index_output_path = "/content/drive/MyDrive/IR_Bonus_Project/phase_4_results/index.faiss"
metadata_output_path = "/content/drive/MyDrive/IR_Bonus_Project/phase_4_results/metadata.json"


document_ids = []
embeddings = []
for i in range(len(contents)):
      data = load_from_json(f'{embeddings_path}{i}.json')
      document_ids.append({"doc_id": data["doc_id"], "content": data["content"]})
      embeddings.append(data["embedding"])


embeddings_matrix = np.array(embeddings).astype('float32')

index = faiss.IndexFlatL2(embeddings_matrix.shape[1])
index.add(embeddings_matrix)
faiss.write_index(index, index_output_path)

save_to_json(document_ids, metadata_output_path)


print(f"FAISS index saved to {index_output_path}")
print(f"Document metadata saved to {metadata_output_path}")
```

```python
queries = ["ویروس کرونا",
           "تهاجم فرهنگی و اسعمار اروپایی ها",
           "ایدئولوژی کمونیستی و سوسیالیستی",
           "انقلاب صنعتی و آغاز تولید ماشین بخار"
          ]
```

```python
import time

def retrieve_documents(query_embedding, index, document_ids, top_k=5):
    start_time = time.time()


    distances, indices = index.search(np.array([query_embedding]).astype('float32'), top_k)


    results = [
        {"doc_id": document_ids[idx]["doc_id"], "content": document_ids[idx]["content"], "score": 1 - (distances[0][i] / (max(distances[0]) + 0.00001 ))}
        for i, idx in enumerate(indices[0]) if idx != -1
    ]

    retrieval_time = time.time() - start_time
    return results, retrieval_time
```

```python
index = faiss.read_index(index_output_path)
data = load_from_json(metadata_output_path)
document_ids = data

for query in queries:
  query_embedding = model.encode(query, convert_to_tensor=False)

  top_k = 5
  results, retrieval_time = retrieve_documents(query_embedding, index, document_ids, top_k)


  print(f"Retrieved top {top_k} documents in {retrieval_time:.4f} seconds for query => \"{query}\" :")
  for rank, result in enumerate(results, start=1):
      print(f"Rank {rank}:")
      print(f"Document ID: {result['doc_id']}")
      print(f"Document Content: {result['content']}")
      # print(f"Score: {result['score']:.7f}\n------------------")
  print("<=======================================================>")
```
